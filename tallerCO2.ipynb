# Media Global Mensual de CO2

usando estos codigos y los datos obtenidos por (Earth System Research Laboratory) analizaremos el  grafico de el promedio de dioxido de carbono presente sobre la superficie marina terrestre.

```python
from pylab import plot, show, xlabel, ylabel
from numpy import loadtxt
dataCO2=loadtxt("co2_mm_gl.txt",float)
plot(dataCO2[:,2],dataCO2[:,3])
plot(dataCO2[:,2], dataCO2[:,4])
xlabel("A침o")
ylabel("Partes por millon")
show()
```

Esta grafica presenta un comportamiento casi lineal con una pendiente positiva es decir, Cada a침o los niveles de CO2 aumentan de manera considerable, este comportamiento se presenta a nivel general, sin embargo a nivel mas local si tomamos en cuenta los datos de los ultimos a침os, podemos ver que la grafica tambien tiene un comportamiento oscilatorio. esto es asi debido a que la media de CO2 decrece debido a la posici칩n del planeta respecto a el sol.

```python
import matplotlib.pyplot as plt
import numpy as np
x=dataCO2[:240,2]
y=dataCO2[:240,3]
plot(dataCO2[:240,2],dataCO2[:240,3])
media_x=x.mean()
media_y=y.mean()
m = np.sum((x - media_x)*(y - media_y)) / np.sum((x - media_x)**2)
b = media_y - m * media_x
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + m * x_vals
xlabel("A침o")
ylabel("Partes por millon")
plt.plot(x_vals, y_vals, color="red")
plt.show()
```
